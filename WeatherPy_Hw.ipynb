```python
#api_key = "da3b5287b648748e4a82eab24c09b006"

data = "/Users/scott/Bootcamp Lessons and Exercises/6.3 Visualizing Financial Deserts with Google Places/citipy/worldcities.csv"
cities_df = pd.read_csv(data)
cities_df.head()
```

```python
cities_500 = cities_df.sample(500)
cities_500.head()
```

```python
row_counter = 0

cities_500["Temperature"] = ""
cities_500["Humidity"] = ""
cities_500["Cloudiness"] = ""
cities_500["Wind Speed"] = ""

for index, row in cities_500.iterrows():
    url = "https://api.openweathermap.org/data/2.5/weather?"
    
    lat=str(row["Latitude"])
    lng=str(row["Longitude"])
    units = "Imperial"
    
    query_url = f"{url}lat={lat}&lon={lng}&appid={api_key}&units={units}"

    #query_url = url+"lat="+str(row["Latitude"])+"&lon="+str(row["Longitude"])+"&appid="+api_key+"&units="+units
    
    print(f"Now retrieving city number {row_counter}")
    print(query_url)
    row_counter = row_counter + 1
    
    cities_weather = req.get(query_url).json()
    
    temperature = cities_weather["main"]["temp"]
    humidity = cities_weather["main"]["humidity"]
    cloudiness = cities_weather["clouds"]["all"]
    wind_speed = cities_weather["wind"]["speed"]
        
    cities_500.set_value(index, 'Temperature', temperature)
    cities_500.set_value(index, "Humidity", humidity)
    cities_500.set_value(index, "Cloudiness", cloudiness)
    cities_500.set_value(index, "Wind Speed", wind_speed)
        
cities_500.head()
```

```python
cities_500.to_csv("WeatherData.csv", index=False)
```
